```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the match
teams = ['Lions FC', 'Eagles United']
possession = [55, 45]
shots = [10, 8]
shots_on_target = [7, 4]
expected_goals = [2.5, 1.1]
goals = [3, 1]

# Data for player performance (Radar chart)
players = ['David Kamau (Lions FC)', 'Paul Okello (Eagles United)']
player_metrics = ['Goals', 'Assists', 'Key Passes', 'Dribbles Completed', 'Tackles']
kamau_performance = [2, 1, 4, 5, 1]
okello_performance = [1, 0, 1, 2, 5]

# --- Bar chart for Shots, Shots on Target, Expected Goals, and Goals ---
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Shots and Shots on Target
axs[0, 0].bar(teams, shots, width=0.4, label='Shots', align='center', color='orange')
axs[0, 0].bar(teams, shots_on_target, width=0.4, label='Shots on Target', align='edge', color='green')
axs[0, 0].set_title('Shots and Shots on Target')
axs[0, 0].legend()

# Expected Goals vs Actual Goals
axs[0, 1].bar(teams, expected_goals, width=0.4, label='Expected Goals', align='center', color='cyan')
axs[0, 1].bar(teams, goals, width=0.4, label='Actual Goals', align='edge', color='magenta')
axs[0, 1].set_title('Expected Goals vs Actual Goals')
axs[0, 1].legend()

# --- Pie chart for Possession ---
axs[1, 0].pie(possession, labels=teams, autopct='%1.1f%%', colors=['blue', 'red'], startangle=90)
axs[1, 0].set_title('Possession (%)')

# --- Radar chart for Player Performance Comparison ---
# Normalize data for radar chart (max values set to 5 for radar chart aesthetics)
labels = np.array(player_metrics)
num_vars = len(labels)

# Create 360-degree angles for radar chart
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop

# Radar chart for David Kamau
fig_radar, ax_radar = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
kamau_values = kamau_performance + kamau_performance[:1]
okello_values = okello_performance + okello_performance[:1]

ax_radar.fill(angles, kamau_values, color='blue', alpha=0.25)
ax_radar.fill(angles, okello_values, color='red', alpha=0.25)
ax_radar.plot(angles, kamau_values, label=players[0], color='blue', linewidth=2)
ax_radar.plot(angles, okello_values, label=players[1], color='red', linewidth=2)

# Add labels and titles
ax_radar.set_title('Player Performance Comparison (Radar Chart)', size=15, color='black', y=1.1)
ax_radar.set_yticklabels([])
ax_radar.set_xticks(angles[:-1])
ax_radar.set_xticklabels(labels)
ax_radar.legend(loc='upper right')

# Adjust layout and show all charts
plt.tight_layout()
plt.show()
```
